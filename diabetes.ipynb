1. Deskripsi Library

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

2. Load Dataset

```python
diabetes_dataset = pd.read_csv('/content/diabetes-dataset/diabetes.csv')
```

```python
#select data
diabetes_dataset.head()
```

```python
# jumlah baris = 768
# jumlah kolom = 9
diabetes_dataset.shape
```

```python
# melihat jumlah data yang terkena diabetes dan tidak terkena diabetes
diabetes_dataset['Outcome'].value_counts()

# terkena diabetes = 268
# tidak diabetes = 500
```

```python
# memisahkan data diagnosa dan data label
x = diabetes_dataset.drop(columns='Outcome', axis=1)
y = diabetes_dataset['Outcome']
```

```python
# print data x
print(x)
```

```python
# print data y
print(y)
```

3. Standarirasi Data (Transform data)

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
# Transform data x

data_standar = scaler.transform(x)
```

```python
print(data_standar)
```

```python
#inisialisasi
x = data_standar
y = diabetes_dataset['Outcome']
```

```python
# cetak data yang sudah di skala
print(x)
print(y)
```

4. Split data training dan data testing

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)

# data testing = 0.2 atau 20%
```

```python
# mencetak jumlah data
# x = 768,
# x train = 614
# x test = 154
print(x.shape, x_train.shape, x_test.shape)
```

5. Membuat data latih menggunakan algoritma svm

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(x_train, y_train)
```

6. Membuat model evalusi untuk mengukur tingkat akurasi

```python
x_train_prediction = model.predict(x_train)
training_data_akurasi = accuracy_score(x_train_prediction, y_train)
```

```python
print('Akurasi data training = ', training_data_akurasi)
```

```python
x_test_prediction = model.predict(x_test)
test_data_akurasi = accuracy_score(x_test_prediction, y_test)
```

```python
print('Akurasi data Testing = ', test_data_akurasi)
```

7. Membuat model prediksi

```python
#Pregnancies = 6,
# Glucose = 148,
# BloodPressure = 72,
# SkinThickness = 35,
# Insulin = 0,
# BMI = 33.6,
# DiabetesPedigreeFunction = 0.627,
# Age = 50
input_data = (6, 148, 72, 35, 0, 33.6, 0.627, 50)
input_data_as_array = np.array(input_data)
input_datareshape = input_data_as_array.reshape(1,-1)

std_data = scaler.transform(input_datareshape)
prediction = model.predict(std_data)
if(prediction[0] == 0) :
    print('Pasien TIDAK terkena DIABETES')
else :
    print('Pasien TERKENA DIABETES')
```
